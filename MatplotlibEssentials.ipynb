```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Plot
x = np.arange(0, 10, 1)
y = x ** 2

plt.plot(x, y, marker="o", color="b", linestyle="--", label="y = x^2")
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Bar Chart
names = ["Alice", "Bob", "Charlie", "David"]
salaries = [50000, 60000, 70000, 80000]

plt.bar(names, salaries, color="orange")
plt.title("Employee Salaries")
plt.xlabel("Employee")
plt.ylabel("Salary")
plt.show()
```

```python
# Horizontal Bar Chart
plt.barh(names, salaries, color="green")
plt.title("Employee Salaries (Horizontal)")
plt.xlabel("Salary")
plt.ylabel("Employee")
plt.show()
```

```python
# Pie Chart
sizes = [40, 30, 20, 10]
labels = ["Python", "Java", "C++", "Others"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Programming Language Popularity")
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)  # 1000 random values

plt.hist(data, bins=20, color="purple", edgecolor="black")
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="red")
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Multiple Plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Multiple Plots Example")
plt.legend()
plt.show()
```

```python
# Subplots
x = np.linspace(0, 10, 100)

plt.subplot(2, 1, 1)  # (rows, cols, index)
plt.plot(x, np.sin(x))
plt.title("Sine Function")

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.title("Cosine Function")

plt.tight_layout()
plt.show()
```
